```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data/generalization_results.csv")

# Create pivot table for heatmap
pivot = df.pivot(index="data_size", columns="params", values="gap")

# Plot
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Generalization Gap Across Model and Dataset Sizes")
plt.xlabel("Model Parameters")
plt.ylabel("Training Dataset Size")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("data/generalization_results.csv")
df["p/n"] = df["params"] / df["data_size"]

# Unique data sizes
data_sizes = sorted(df["data_size"].unique())

# Plot one graph per data_size
for n in data_sizes:
    sub = df[df["data_size"] == n].sort_values("params")
    plt.figure(figsize=(8, 5))
    plt.plot(sub["params"], sub["gap"], marker='o', label=f"n = {n}")
    plt.title(f"Generalization Gap vs Model Size for Dataset Size = {n}")
    plt.xlabel("Model Parameters (p)")
    plt.ylabel("Generalization Gap")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both datasets
df_fashion = pd.read_csv("data/generalization_results_fashion_mnist.csv")
df_cifar = pd.read_csv("data/generalization_results.csv")

# Compute p/n ratios
df_fashion["p/n"] = df_fashion["params"] / df_fashion["data_size"]
df_cifar["p/n"] = df_cifar["params"] / df_cifar["data_size"]

# Start plot
plt.figure(figsize=(10, 6))

# FashionMNIST curves
for n in sorted(df_fashion["data_size"].unique()):
    sub = df_fashion[df_fashion["data_size"] == n].sort_values("params")
    plt.plot(sub["params"], sub["gap"], marker='o', linestyle='-', label=f"FashionMNIST n={n}")

# CIFAR-10 curves
for n in sorted(df_cifar["data_size"].unique()):
    sub = df_cifar[df_cifar["data_size"] == n].sort_values("params")
    plt.plot(sub["params"], sub["gap"], marker='x', linestyle='--', label=f"CIFAR-10 n={n}")

# Labels and layout
plt.title("Generalization Gap vs Model Size\nAcross FashionMNIST and CIFAR-10")
plt.xlabel("Model Parameters (p)")
plt.ylabel("Generalization Gap")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
